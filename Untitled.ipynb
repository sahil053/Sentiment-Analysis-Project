```python
import ast
```

```python
def func(obj):
    genre = []
    for dictionary in ast.literal_eval(obj):
        genre.append(dictionary['name'])
    return genre
```

```python
def cast(obj):
    cast = []
    count = 0 
    for dictionary in ast.literal_eval(obj):
        if count!=3:
            cast.append(dictionary['name'])
            count = count + 1
        else:
            break
    return cast
```

```python
df['cast'] = df['cast'].apply(cast)
```

```python
def crew(object):
    L = []
    for dict1 in ast.literal_eval(object):
        if dict1['job'] == 'Director':
            L.append(dict1['name'])
            break
    return L 
```

```python
df['crew'] = df['crew'].apply(crew)
```

```python
#stemming and stopwords removal 
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def preprocessing(text):
    x = []
    for word in text.split():
        if word in stopwords.words('english'):
            x.append('')
        else:
            x.append(ps.stem(word))
    return " ".join(x)
```

```python
preprocessing(df['tags'][0])
```

```python
df['tags'][0]
```

```python
df['tags'] = df['tags'].apply(preprocessing)
```

```python
#Modelling 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer(max_features=5000)
```

```python
opvectors = tf.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(opvectors)
```

```python
similarity.shape #4806 movie similarity with 4806 movies 
```

```python
df.head()
```

```python
df[df['title']=='Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True , key=lambda x:x[1])[1:6]
```

```python
def recommend(title):
    movie_index = df[df['title']==title].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True , key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df['title'][i[0]])
```

```python
recommend('Batman Begins')
```
